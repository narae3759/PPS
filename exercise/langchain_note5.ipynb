```python
from langchain_community.document_loaders import AsyncChromiumLoader

# 크롤링할 URL 목록을 설정합니다.
urls = ["https://n.news.naver.com/mnews/article/138/0002172287"]

# AsyncChromiumLoader를 사용하여 URL에서 비동기적으로 문서를 로드합니다.
loader = AsyncChromiumLoader(urls)
# 로드된 문서를 가져옵니다.
html = loader.load()
print(html[0].page_content)
```

```python
from langchain_community.document_transformers import Html2TextTransformer
# pip install html2text
html2text = Html2TextTransformer()
html_transformed = html2text.transform_documents(html)
print(html_transformed[0].page_content)
```

```python
from langchain_community.document_transformers import BeautifulSoupTransformer
# pip install html2text
html2text = BeautifulSoupTransformer()
html_transformed = html2text.transform_documents(html, tags_to_extract=["a"])
print(html_transformed[0].page_content)
```

```python
from langchain_community.document_loaders import WebBaseLoader

# 크롤링할 URL 목록을 설정합니다.
urls = ["https://n.news.naver.com/mnews/article/138/0002172287"]

# AsyncChromiumLoader를 사용하여 URL에서 비동기적으로 문서를 로드합니다.
loader = WebBaseLoader(urls, encoding="utf-8")
# 로드된 문서를 가져옵니다.
docs = loader.load()
print(docs[0])
```
